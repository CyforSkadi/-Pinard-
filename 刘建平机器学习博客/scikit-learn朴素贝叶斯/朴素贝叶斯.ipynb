### 先验为高斯分布

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
Y = np.array([1, 1, 1, 2, 2, 2])
clf = GaussianNB()

clf.fit(X,Y)
```

```python
clf.predict([[-0.8,-1]])
```

```python
clf.predict_proba([[-0.8,-1]])
```

```python
clf.predict_log_proba([[-0.8,-1]])
```

### 先验为多项式分布

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

X1,Y1 = make_classification(n_samples=400,n_features=2,n_redundant=0,n_clusters_per_class=1,n_classes=3)
plt.scatter(X1[:,0],X1[:,1],c=Y1)
plt.show()
```

```python
clf = MultinomialNB(alpha=1,fit_prior=True)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X1)
Y = Y1

clf.fit(X,Y)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
clf.predict_proba([[0.2,0.2]])
```

```python
clf.predict_proba([[0.8,0.8]])
```
